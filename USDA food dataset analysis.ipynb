```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
```

```python
db = pd.read_json('USDA_food.json')
```

```python
db
```

```python
nutrients = []
for row in range(6636):
    fnut = DataFrame(db['nutrients'][row])
    fnut['id'] = db['id'][row]
    nutrients.append(fnut)
nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients
```

```python
nutrients.duplicated().sum()
```

```python
 nutrients = nutrients.drop_duplicates()
```

```python
nutrients
```

```python
db = db.drop(['nutrients','portions','tags', 'manufacturer'], axis = 1)
```

```python
db
```

```python
 col_mapping = {'description' : 'food',
 'group' : 'fgroup'}
db = db.rename(columns=col_mapping, copy=False)
```

```python
 col_mapping = {'description' : 'nutrient',
 'group' : 'nutgroup'}
nutrients = nutrients.rename(columns=col_mapping, copy=False)
```

```python
db
```

```python
nutrients
```

```python
ndata = pd.merge(nutrients, db, on='id', how='outer')
```

```python
ndata
```

```python
result = ndata.groupby(['nutrient', 'fgroup'])['value'].quantile(0.5)
result
```

```python
graph = result['Zinc, Zn'].sort_values().plot(kind = 'bar')
```
